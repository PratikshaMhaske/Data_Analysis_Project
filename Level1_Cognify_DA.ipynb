# **TASK 1 - Top Cuisines**

# **Top 3 Cuisine**

```python
# Split multiple cuisines into individual ones
all_cuisines = df['Cuisines'].dropna().str.split(', ').explode()

# Count individual cuisines
top_3_cuisines = all_cuisines.value_counts().head(3)

top_3_cuisines
```

Initially, value_counts was used which counted cuisine combinations.
To correctly identify popular cuisines, the cuisines column was split into individual cuisines and then analyzed.

# **Percentage Calculation**

```python
total_restaurants = df['Cuisines'].dropna().shape[0]

percentage = (top_3_cuisines / total_restaurants) * 100

result = pd.DataFrame({
    'Cuisine': top_3_cuisines.index,
    'Count': top_3_cuisines.values,
    'Percentage (%)': percentage.round(2)
})

result
```

# **TASK 2 - City Analysis**

# **City with Highest Number of Restaurants**

```python
# Count restaurants per city
city_counts = df['City'].value_counts()

# City with highest number of restaurants
top_city = city_counts.idxmax()
top_city, city_counts[top_city]
```

# **Average Rating for Each City**

```python
# Remove rows with missing ratings
city_rating = df.dropna(subset=['Aggregate rating'])

# Calculate average rating per city
avg_rating_city = city_rating.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)

avg_rating_city.head()
```

# **City with Highest Average Rating**

```python
highest_avg_rating_city = avg_rating_city.idxmax()
highest_avg_rating = avg_rating_city.max()

highest_avg_rating_city, highest_avg_rating
```

# **Visualization**

```python
import matplotlib.pyplot as plt

city_counts.head(10).plot(kind='bar')
plt.title('Top 10 Cities by Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()
```

The city-wise analysis was performed to understand the distribution of restaurants across cities.
The city with the highest number of restaurants was identified using value counts.
Average restaurant ratings were calculated for each city to evaluate performance, and the city with the highest average rating was determined.

# **TASK 3 - Price Range Distribution**

```python
df['Price range'].value_counts().sort_index()
```

# **Visualization**

```python
import matplotlib.pyplot as plt

price_counts = df['Price range'].value_counts().sort_index()

price_counts.plot(kind='bar')
plt.title('Distribution of Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.show()
```

# **Percentage Calculation**

```python
total_restaurants = df.shape[0]

price_percentage = (price_counts / total_restaurants) * 100

price_result = pd.DataFrame({
    'Price Range': price_counts.index,
    'Count': price_counts.values,
    'Percentage (%)': price_percentage.round(2)
})

price_result
```

The distribution of restaurant price ranges was analyzed to understand affordability trends. A bar chart was used to visualize the number of restaurants in each price range category. The percentage of restaurants in each category was also calculated to identify the dominant pricing segment.

# **TASK 4 - Online Delivery**

```python
df['Has Online delivery'].value_counts()
```

# **Percentage of Restaurants Offering Online Delivery**

```python
online_counts = df['Has Online delivery'].value_counts()

online_percentage = (online_counts / df.shape[0]) * 100

online_result = pd.DataFrame({
    'Online Delivery': online_counts.index,
    'Count': online_counts.values,
    'Percentage (%)': online_percentage.round(2)
})

online_result
```

# **Compare Average Ratings**

```python
# Remove missing ratings
delivery_rating = df.dropna(subset=['Aggregate rating'])

# Average rating based on online delivery
avg_rating_delivery = delivery_rating.groupby('Has Online delivery')['Aggregate rating'].mean()

avg_rating_delivery
```

# **Visualization**

```python
avg_rating_delivery.plot(kind='bar')
plt.title('Average Rating: Online Delivery vs No Online Delivery')
plt.xlabel('Online Delivery')
plt.ylabel('Average Rating')
plt.show()
```

This analysis evaluates the impact of online delivery on restaurant ratings. The percentage of restaurants offering online delivery was calculated, and the average ratings of restaurants with and without online delivery were compared to understand customer preferences.
